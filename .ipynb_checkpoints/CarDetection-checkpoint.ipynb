### Car Detection

```python
# import for traiing data import
import glob

# import for matplotlib
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

import numpy as np
import cv2
import random

# imports for video data
from moviepy.editor import VideoFileClip
from IPython.display import HTML
%matplotlib inline
```

### Read in data images

```python

# read in training data
vehicle_images_location = glob.glob('./data/vehicles/*/*.png')
non_vehicle_images_location = glob.glob('./data/non-vehicles/*/*.png')
```

```python
print ("Number of samples of Vehicles: {0}\nNumber of samples of Non-Vehicles: {1}".format(len(vehicle_images_location), len(non_vehicle_images_location)))
```

```python




vehicle_images = []
non_vehicle_images = []

fig = plt.figure(figsize=(15,7))

for x in range(10):
    # genreate random numbers to get random images
    rand_number = random.randint(1,len(vehicle_images_location))
    
    plt.subplot(2,5,x+1)
    # read in image based on random number
    vehicle = cv2.imread(vehicle_images_location[rand_number])
    vehicle = cv2.cvtColor(vehicle,cv2.COLOR_BGR2RGB)
    # append image to array to use later
    vehicle_images.append(vehicle_images_location[rand_number])
    plt.imshow(vehicle)
    plt.axis('Off')
    # slice up path - only file name
    img_index = vehicle_images_location[rand_number].split('/')[-1]
    # create a plot title
    title = "Vehicle Image\n File: {0}".format(img_index)
    # add custom title to each plot
    plt.title(title, wrap=True)
plt.show()

fig = plt.figure(figsize=(15,7))
                   
for x in range(10):
    # genreate random numbers to get random images
    rand_number = random.randint(1,len(non_vehicle_images_location))
    
    plt.subplot(2,5,x+1)
    # read in image based on random number
    non_vehicle = cv2.imread(non_vehicle_images_location[rand_number])
    non_vehicle = cv2.cvtColor(non_vehicle,cv2.COLOR_BGR2RGB)
    # append image to array to use later
    non_vehicle_images.append(non_vehicle_images_location[rand_number])
    plt.imshow(non_vehicle)
    plt.axis('Off')
    # slice up path - only file name
    img_index = non_vehicle_images_location[rand_number].split('/')[-1]
    # create a plot title
    title = "Non-Vehicle Image\n File: {0}".format(img_index)
    # add custom title to each plot
    plt.title(title, wrap=True)
plt.show()
```

### Hog Feature Detection

```python
def get_hog_features(img, orient, pix_per_cell, cell_per_block, 
                        vis=False, feature_vec=True):
    # Call with two outputs if vis==True
    if vis == True:
        features, hog_image = hog(img, orientations=orient, 
                                  pixels_per_cell=(pix_per_cell, pix_per_cell),
                                  cells_per_block=(cell_per_block, cell_per_block),
                                  block_norm= 'L2-Hys',
                                  transform_sqrt=False, 
                                  visualise=vis, feature_vector=feature_vec)
        return features, hog_image
    # Otherwise call with one output
    else:      
        features = hog(img, orientations=orient, 
                       pixels_per_cell=(pix_per_cell, pix_per_cell),
                       cells_per_block=(cell_per_block, cell_per_block),
                       block_norm= 'L2-Hys',
                       transform_sqrt=False, 
                       visualise=vis, feature_vector=feature_vec)
        return features
```
